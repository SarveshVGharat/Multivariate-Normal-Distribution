# Testing Functions (Jax)

```python
import math
import matplotlib.pyplot as plt
import multivariate_jax as mu
import jax
from jax import random
import numpy as np
```

```python
d = 10
key = random.PRNGKey(42)
mean = random.uniform(key, shape=(1,d)).T
random_matrix = random.uniform(key, shape=(d,d))
covariance = jax.numpy.matmul(random_matrix, random_matrix.transpose())
```

```python
mu.box_muller(n, d)
```

```python
mu.cholesky_decomposition(n, d, mean, covariance)
```

```python
mu.mean_cov_diff(n, d, mean, covariance)
```

```python
mu.plot_mean_cov_diff(n, d, mean, covariance)
```
